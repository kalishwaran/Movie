# Pre- Processing

```python
import ast
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
    
```

```python
df['genres']=df['genres'].apply(convert)
```

```python
df.head(10)
```

```python
df['keywords'][0]
```

```python
df['keywords']=df['keywords'].apply(convert)
```

```python
df['keywords'][0]
```

```python
df.head(10)
```

```python
df['cast'][0]
```

```python
df['cast']=df['cast'].apply(convert).apply(lambda x: x[:3])
```

```python
df['cast'][0]
```

```python
df.head(10)
```

```python
df['crew'][4]
```

```python
def director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
df['crew']=df['crew'].apply(director)
```

```python
df['crew'][0]
```

```python
def space (L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
df['genres']=df['genres'].apply(space)
df['keywords']=df['keywords'].apply(space)
df['cast']=df['cast'].apply(space)
df['crew']=df['crew'].apply(space)
```

```python
df.head()
```

```python
df['cast'][0]
```

```python
df['crew'][0]
```

```python
df['overview']=df['overview'].apply(lambda x : x.split())
df['tag']=df['cast']+df['crew']+df['keywords']+df['genres']+df['overview']
new=df.drop(columns=['genres','cast','crew','overview','keywords'])
```

```python
new['tag']=new['tag'].apply(lambda x : " ".join(x))
```

```python
new.head(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new['tag']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cs=cosine_similarity(vector)
```

```python
cs
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(cs[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
    
```

```python
recommend('Avatar')
```

```python
recommend('Batman v Superman: Dawn of Justice')
```

```python
recommend('Harry Potter and the Half-Blood Prince')
```
